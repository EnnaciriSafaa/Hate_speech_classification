# 1. Importing Libraries

```python
import pandas as pd
import numpy as np
import pickle, time

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

from sklearn.pipeline import Pipeline, make_pipeline
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score, precision_recall_fscore_support, accuracy_score
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

# 2. Data Preparation 

## 2.1 Load and explore data

```python
# Load training data
df = pd.read_csv('../../Data/train_set1.csv')
df.head()
```

```python
# Load testing data
test = pd.read_csv('../../Data/test_set1.csv')
test.head()
```

```python
X_train = df['comment_text']
Y_train = df['tox']

#Check shape of data
print(X_train.shape)
print(Y_train.shape)
```

```python
X_test = test['comment_text']
Y_test = test['tox']

#Check shape of data
print(X_test.shape)
print(Y_test.shape)
```

# 3. Classifier training

## 3.1 Create pipeline

```python
start_time=time.time()

SVC_pipeline1 = make_pipeline(TfidfVectorizer(),
                SVC(verbose=True))


param_grid = {'svc__kernel': ['linear','sigmoid','rbf', 'poly'], 
              'svc__gamma': ['auto', 'scale']
             } 

            
grid_3 = GridSearchCV(SVC_pipeline1, param_grid, cv=10, scoring='accuracy', error_score='raise')

grid = grid_3 .fit(X_train, Y_train)

end_time=time.time()
              
print("total time",end_time-start_time)
```

```python
# Print the model parameters of the best estimator
print(grid.best_estimator_.named_steps['svc'])
```

```python
# Print the model parameters of the best model
grid.best_params_
```

```python
# Print the training score of the best model
print('Test accuracy: %.3f' % grid.best_score_)
```

```python
# Print the test score of the best model
best_estimator = grid.best_estimator_
print('Test accuracy: %.3f' % best_estimator.score(X_test, Y_test))
```
