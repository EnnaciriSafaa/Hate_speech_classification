# 1. Importing Libraries

```python
import pandas as pd
import numpy as np
import pickle, time

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score, precision_recall_fscore_support, accuracy_score
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.model_selection import GridSearchCV
```

# 2. Data Preparation 

## 2.1 Load and explore data

```python
# Load training data
df = pd.read_csv('../../Data/train_set4.csv')
df.head()
```

```python
X_train = df['text']
Y_train = df['tox']

#Check shape of data
print(X_train.shape)
print(Y_train.shape)
```

# 3. Classifier training

## 3.1 Create pipeline

```python
start_time=time.time()

# Create the parameter grid
param_grid = [{'logisticregression__penalty': ['l1', 'l2'], # Regularization penalty 
               'logisticregression__C' : [0.01, 0.1, 1, 10], #  Inverse of regularization strength
               'logisticregression__class_weight' : ['balanced', None], # Weights associated with classes
               'logisticregression__solver' : ['liblinear', 'saga']}] # Solvers for optimization  
 

# Create a pipeline
LR_pipeline = make_pipeline(TfidfVectorizer(), 
                             LogisticRegression(max_iter=1000, random_state=0, verbose=True))

# Create an instance of GridSearch Cross-validation estimator
grid_pipeline = GridSearchCV(LR_pipeline, param_grid, cv=10, scoring='accuracy', error_score='raise')


# Train the svm
grid = grid_pipeline .fit(X_train, Y_train)

end_time=time.time() 

# print time of execution
print("total time",end_time-start_time)
```

```python
print(grid.best_estimator_.named_steps['logisticregression'])
```
