Create Excel file to save results's evaluation metrics of JgsawToxic set 

```python
import openpyxl 
from openpyxl import Workbook
from openpyxl.styles import Alignment
from openpyxl.styles.borders import Border, Side
```

```python
workbook = Workbook()

# grab the active worksheet
worksheet = workbook.active

from string import ascii_uppercase

newFile = "results_LR_set3.xlsx"

# define column dimensions
for column in ascii_uppercase:
    worksheet.column_dimensions[column].width = 25
    
# merge cells    
worksheet.merge_cells('B1:E1')
worksheet.merge_cells('F1:I1')
worksheet.merge_cells('A1:A2')

worksheet['A1']= 'Approaches'
worksheet['B1']= 'Train data'
worksheet['F1']= 'Test data'

# Rows can also be appended
worksheet.append(['Approaches', 'train_Accuracy', 'train_Recall', 'train_F1-mesure', 'train_Precision', 
                  'test_Accuracy', 'test_Recall', 'test_F1-mesure', 'test_Precision'])

# define style border
thin_border = Border(left=Side(style='thin'), 
                     right=Side(style='thin'), 
                     top=Side(style='thin'), 
                     bottom=Side(style='thin'))

cell_alignment = Alignment(horizontal='center')

# apply style border
for row in worksheet:
    for cell in row:
        cell.border = thin_border
        cell.alignment = cell_alignment
        
# Save the file
workbook.save(newFile)
```
